```python
#HOMEWORK3- GÜLÇİN SÖNMEZ

#5 students were entered in nested dictionary, data record was created and student information is included
students={
    "0":{
        "username":"Ali MEŞE"
    },
    "1":{
        "username":"Veli YOGURTÇU"
    },
    "2":{
        "username":"Selin MERT"
    },
    "3":{
        "username":"Jonny Depp"
    },
    "4":{
        "username":"Melisa OZ"
    },
    
}

passList=[]  #A list has been created to include all passing grades.


for x in range(5):
    student_number=str(x)
    
    print("..............Course Grade Application...............\n")  
   

   # username = input("Name: " )  #The name of the student whom the user wanted to add her notes to was asked.
    
    
    print(students[student_number]["username"],"'s information:")
    
 #Grade information was asked to the user
    midterm=int(input("Midterm= "))
    project_grade= int(input("Project Grade= "))
    final_grade= int(input("Final Grade= "))
    passing_grade=int((midterm * 0.3 + project_grade* 0.3 + final_grade * 0.4 ))   #Passing grade = midterm exam (0.3) + project (0.3) + final * (0.4) passing grade should be determined in this way.
    print("Passing Grade= ", passing_grade)

  #added to the information dictionary  
    student=students[student_number]
    student["midterm"]=midterm
    student["project grade"]=project_grade
    student["final grade"]=final_grade


  #passing grade is set as 60
    if (passing_grade >= 60 ):
        
        student["passing grade"]=passing_grade,"BAŞARILI"
        passList.append(student["passing grade"][0])   #Passing notes have been added to the list.
    
    else:
        student["passing grade"]=passing_grade,"BAŞARISIZ"
        passList.append(student["passing grade"][0])  #Passing notes have been added to the list.
        
        
    print(student["username"],"'s information:",student)
    passList.sort() #the list is sorted
    passList.reverse() #passing notes are ranked from higher to lower 
    
    print("-"*50,"\n")
    print("\nAll passlist: ",passList,"\n\n")
    
    
```
