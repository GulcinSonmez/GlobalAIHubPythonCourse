```python
#HOMEWORK_1
#Gülçin_Sönmez

mylist = list(range(31)) #Any range 
print("Mylist:", mylist)

even= list(filter(lambda x: (x % 2 == 0), mylist)) #first list
print("\nEven number's: ", even)

odd=list(filter(lambda x:(x % 2 == 1), mylist)) #second list
print("\nOdd number's: ", odd)

#merge these two list
combination =odd+even 
combination.sort()
print("\nCombination of even and odd numbers lists:",combination)

#multiply all values in the new list by 2 and print data type of all values in the new list
for i in combination:
        nmbr=i*2 
        print("\n",nmbr, type(nmbr)) 
```
